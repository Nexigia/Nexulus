```python
#그래서 어떻게 series -> DataFrame으로 바꿀까 하다가 그런 메소드가 있나 해서
# GPT한테 물어봤더니 to_frame()을 알려주더군요

brand_counts = car_df['brand'].value_counts()
brand_avg_price = car_df.groupby('brand')['Price'].mean()

brand_counts.to_frame(name='count').join(brand_avg_price.to_frame(name='avg_price'))
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
```

```python
car_rs_df = rs.fit_transform(car_df)
car_df = pd.DataFrame(car_rs_df, columns=car_df.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(5)
```

```python
pca_result = pca.fit(car_df)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
for i in range(2, 11):
    pca = PCA(i)
    pca.fit(car_df)
    print(i, round(pca.explained_variance_ratio_.sum(), 2))
```
